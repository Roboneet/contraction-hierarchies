## Original Graph

```python
import json
import jgraph
```

```python
data = json.load(open('graph.json'))
```

```python
print(json.dumps(data, indent=2))
```

```python
ee = [map(int, data["edges"][i]["nodes"]) for i in data["edges"]]
```

```python
Edges = [ (ee[i][0], ee[i][1]) for i in range(len(ee))]
```

```python
jgraph.draw(Edges, directed=False)
```

# Contracted Graph

```python
contractedData = json.load(open('contractedGraph.json'))
```

```python
kk = [map(int, contractedData["edges"][i]["nodes"]) for i in contractedData["edges"]]
```

```python
ContEdges = [ (kk[i][0], kk[i][1]) for i in range(len(kk))]
```

```python
jgraph.draw(ContEdges, directed=False)
```

### EDGE DIFFRENCE

```python
len(ContEdges) - len(Edges)
```
